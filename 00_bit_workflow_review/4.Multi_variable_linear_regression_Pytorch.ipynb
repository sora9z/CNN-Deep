```python
# Lab 4 Multi-Variable linear regression

import torch
import torch.nn as nn
from torch.autograd import Variable
import numpy as np

torch.manual_seed(000)

xy=np.loadtxt('data-01-test-score.csv',delimiter=',',dtype=np.float32)
x_data=xy[:,0:-1]
y_data=xy[:,[-1]]

print('x_data',x_data)
print('y_data',y_data)

# numpy를 pytorch 변수로 변환
x_data=Variable(torch.from_numpy(x_data))
y_data=Variable(torch.from_numpy(y_data))

# Our hypothesis
model=nn.Linear(3,1,bias=True) #tf.keras.Sequential() + tf.model.add(tf.keras.Dense(unit=1,input_dim=3,activation='linear')

#cost
criterion=nn.MSELoss()

#Minimize
optimizer=torch.optim.SGD(model.parameters(),lr=1e-5)

#Train the model
for step in range(2001):
    hypothesis=model(x_data) # hypothesis
    cost=criterion(hypothesis,y_data) #cost function
    optimizer.zero_grad() # 기울기 값 0으로 초기화
    cost.backward() # 편미분
    optimizer.step() # update

    if step%10==0:
        print(step,"Cost",cost.data.numpy(),'\nPrediction',hypothesis.data.numpy())

# Ask My score
print("Your Score will be",model(torch.Tensor([[100,70,101]])))
print("Other Score will be",model(torch.Tensor([[60,70,110],[90,100,80]])).data.numpy())


```
