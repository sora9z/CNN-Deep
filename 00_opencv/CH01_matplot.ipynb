## 7강 Matplotlib 사용하여 영상 출력하기


- Matplotlib 라이브러리 
    - 함수 그래프, 차트,히스토그램 등의 다양한 그리기 기능을 제공하는 Python 패키지
    pip install Matplotlib

- 컬러 영상 출력
    - 컬러 영상의 색상 정보가 RGB순서이어야 함.
    - cv2.imread() 함수로 불러온 영상의 색상 정보는 BGR순서이므로 이를 RGB순서로 변경해야함. -> cv2.cvtColor()함수 이용. converter color

- 그레이스케일 영상 출력
    - plt.imshow() 함수에서 컬러맵을 cmap='gray'으로 지정 cmap = color map  의 약자



```python
import matplotlib.pyplot as plt
import cv2

#컬러 영상 출력
imgBGR=cv2.imread('cat.bmp')
imgRGB=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)

plt.axis('off') # 가로 세로 눈금 없애줌.
plt.imshow(imgRGB)
#plt.imshow(imgBGR) # Red 와 Blue 가 바뀌어 출력됨.
#plt.imshow(cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)) #이렇게 써도 가능
plt.show()



     
```

```python
#그레이 스케일 영상 출력
imgGray=cv2.imread('cat.bmp',cv2.IMREAD_GRAYSCALE) #2차원 형태로 저장이 되고, 밝기값만 저장이 된다.

plt.axis('off')
plt.imshow(imgGray,cmap='gray') #color map을 gray로 설정해야 한다.
#plt.imshow(imgGray)
plt.show()
```

```python
# 두 개의 영상을 함께 출력
plt.subplot(121),plt.axis('off'),plt.imshow(imgRGB)
plt.subplot(122),plt.axis('off'),plt.imshow(imgGray,cmap='gray')
#121 : 1개의 행에 표현 , 2개의 열로 나눔, 그 중 1번째 열에 그림을 그려라. 
#122 : 1개의 행에 표현, 2개의 열로 나눔, 그 중 2번째 열에 그림을 그려라.
plt.show()
## 여러가지 영상을 한 와면에 보여주고 싶을때 matplotlib을 사용하면 유용하다.
```
