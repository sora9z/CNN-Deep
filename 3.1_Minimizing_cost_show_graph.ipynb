## 3. Linear Regression cost 함수 최소화
- Cost Function : hypothesis의 성능을 측정하는 것이 목적. Parameters값이 우리가 풀고자 하는 목적에 적합한지 측정하기 위해 손실함수를 정의한다.
- Cost Function을 최소화 할 수 있는 학습 알고리즘 설계 : Gradient Descent가 많이 사용됨.

## 실습 목적
- Cost 함수의 특성을 이해한다.
- Cost 함수의 동작원리를 이해하기 위해0 tensorflow API를 사용하지 않고 Cost 함수 최소화 기능을 구현한다.
- matplotlib 라이브러리를 통하여 결과를 그래픽으로 처리하는 방법을 이해한다.


```python
# Lab 3-1 Minimizing Cost show graph
import tensorflow as tf
import matplotlib.pyplot as plt

tf.set_random_seed(999)

X=[1,2,3]
Y=[1,2,3]

W=tf.placeholder(tf.float32)

hypothesis=X*W

cost=tf.reduce_mean(tf.square(hypothesis-Y))

# Launch Graph session

sess=tf.Session()

# Variables for plotting cost function
W_history=[]
cost_history=[]

for i in range(-30,50):
    curr_W=i*0.1
    curr_cost=sess.run(cost,feed_dict={W:curr_W})
    W_history.append(curr_W)
    cost_history.append(curr_cost)

#Show the cost function
plt.plot(W_history,cost_history)
plt.show()
```
