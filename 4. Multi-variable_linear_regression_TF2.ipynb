```python
# Lab 4 Multi-variable linar regression TF2

import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
from tensorflow.keras.optimizers import SGD


xy=np.loadtxt('data-01-test-score.csv',delimiter=',',dtype=np.float32)
x_data=xy[:,0:-1]
y_data=xy[:,[-1]]

# Make sure the shape and data are OK
print(x_data.shape)
print(y_data.shape)

```

```python
tf.model=Sequential()

# activation function dosen't have to added as a separate layer. 
# Added it as an argument of Dense() layer
tf.model.add(Dense(units=1,input_dim=3,activation='linear'))
# tf.model.add(tf.keras.layer.Activation('linear'))
tf.model.summary()

tf.model.compile(loss='mse',optimizer=SGD(lr=1e-5))
history=tf.model.fit(x_data,y_data,epochs=2000)

```

```python
# Ask my score
print('Your score will be',tf.model.predict([[100,70,101]]))
print('Other score will be',tf.model.predict([[60,70,110],[90,100,80]]))
```

```python
## 파일로부터 data set 불러오기 복습

import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

xy=np.loadtxt('data-01-test-score.csv',delimiter=',',dtype=np.float32)

x_data=xy[:,0:-1]
y_data=xy[:,[-1]]
# Check the data are OK
print('x_data\n',x_data.shape,'\n',x_data)
print('y_data\n',y_data.shape,'\n',y_data)

# Set model
tf.model=Sequential()
tf.model.add(Dense(units=1,input_dim=3,activation='linear'))
tf.model.summary()

# Compile
tf.model.compile(loss='mse',optimizer=tf.keras.optimizers.SGD(lr=1e-5))

history=tf.model.fit(x_data,y_data,epochs=2000)



```

Dense implements the operation: output = activation(dot(input, kernel) + bias) where activation is the element-wise activation function passed as the activation argument, kernel is a weights matrix created by the layer, and bias is a bias vector created by the layer (only applicable if use_bias is True).
