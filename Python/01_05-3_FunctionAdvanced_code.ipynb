```python
## (2) Lambda-함수의 매개변수로 함수 전달하기

# 매개변수로 받은 함수를 10번 호출하는 함수
def call_10_times(func):
    for i in range(10):
        func()

# 간단한 출력하는 함수
def print_hello():
    print('Hello')

# 조합하기
call_10_times(print_hello) #call_10_times는 매개변수로 함수를 전달한다.
```

```python
## (2) Lambda-함수를 매개변수로 전달하는 대표적인 함수
## map() 함수와 filter()함수
## map(Func,List) : 리스트의 요소를 함수에 넣고 리턴된 값으로 새로운 리스트를 구성해 주는 함수
## filter(Func,List) : 리스트의 요소를 함수에 넣고 리턴된 값이 True인 것으로, 새로운 리스트를 구성해 주는 함수

# 함수를 선언한다.
def power(item): # 값을 제곱한 값을 리털한다
    return item*item
def under_3(item): 
    return item<3 # 3보다 작으면 값을 리턴한다.

# 변수를 선언한다.
list_input_a=[1,2,3,4,5]

# map() 함수를 사용한다.
output_a=map(power,list_input_a)
print("# map() 함수의 실행결과")
print("map(power,list_input_a):",output_a)
print("map(power,list_input_a):",list(output_a))
print()

# filter() 함수를 사용한다.
output_b=filter(under_3,list_input_a)
print("# filter() 함수의 실행결과")
print("filter(under_3,list_input_a):",output_b)
print("filter(under_3,list_input_a):",list(output_b))


```

```python
## (2) Lambda Lambda(람다) : 간단한 함수를 쉽게 선언하는 방법
## lambda 매개변수:리턴값
## def를 사용하여 선언하지 않아도 된다.
## 위의 power() 함수와 under_3() 함수를 람다로 변환해보자.

# 함수를 선언한다.
power=lambda x:x*x

under_3=lambda x:x<3

# 변수를 선언한다
list_input_a=[1,2,3,4,5]

# map() 함수를 사용한다.
output_a=map(power,list_input_a)
print("# map() 함수의 실행결과")
print("map(power,list_input_a):",output_a)
print("map(power,list_input_a):",list(output_a))
print()

# filter() 함수를 사용한다.
output_b=filter(under_3,list_input_a)
print("# filger() 함수의 실행결과")
print("filter(under_3,list_input_a):",output_b)
print("filter(under_3,list_input_a):",list(output_b))
print()

```

```python
## (2) Lambda-인라인 람다
## 람다는 함수의 매개변수에 곧바로 넣을 수 있다.
## 코드가 굉장히 깔끔해짐
## 매개변수로 넣어진 함수를 찾지 않아도 됨.
## 매개변수가 여러개인 람다도 생성 가능.

# 변수를 선언한다.
list_input_a=[1,2,3,4,5]

# map() 함수를 사용한다.
output_a=map(lambda x:x*x,list_input_a)
print("# map() 함수의 실행결과")
print("map(power,list_input_a):",output_a)
print("map(power,list_input_a):",list(output_a))
print()

# filter() 함수를 사용한다.
output_b=filter(lambda x:x<3,list_input_a)
print("# filter() 함수의 실행결과")
print("filter(under_3,list_input_a):",output_b)
print("filter(under_3,list_input_a):",list(output_b))
print()
```

```python
## (3) 파일처리-텍스트 한 줄씩 읽기
## 데이터 구정 : 랜덤하게 1000명의 키와 몸무게 만들기

# 랜텀한 숫자를 만들기 위해 가져온다.
import random
# 간단한 한글 리스크를 만든다.
hanguls=list("가나라다마바사아자차카타파하")
# 파일을 쓰기 모드로 연다.
with open("./00_test/info.txt","w") as file:
    for i in range(1000):
        # 랜덤한 값으로 변수를 생성한다.
        name=random.choice(hanguls)+random.choice(hanguls)
        weight=random.randrange(40,100)
        height=random.randrange(140,200)
        # 텍스트를 쓴다.
        file.write("{},{},{}\n".format(name,weight,height))
```

```python
## (3) 파일처리-반복문으로 파일 한 줄씩 읽기
## 파일을 처리하는 코드
## 한 줄씩 읽으면서 키와 몸무게로 BMI를 계산하는 코드 작성 
with open("./00_test/info.txt","r") as file:
    for line in file:
        # 변수를 선언한다.
        (name, weight, height)=line.strip().split(",")

        # 데이터가 문제없는지 확인한다. : 문제가 있으면 지나감
        if (not name) or (not weight) or (not height):
            continue
        # 결과를 계산한다.
        bmi=int(weight)/int((int(height)/100)**2)
        result=""
        if 25<=bmi:
            result="과체중"
        elif 18.5<=bmi:
            result="정상 체중"
        else:
            result="저체중"
        
        # 출력 합니다.
        
       
        print('\n'.join([
            "이름: {}",
            "몸무게 :{}",
            "키 :{}",
            "BMI :{}",
            "결과 :{}"
        ]).format(name,weight,height,bmi,result))
        print()
```

```python
## (5) 좀 더 알아보기-제너레이터(Generator)
## 제너레이터 함수

# 함수를 선언한다.
def test():
    print("힘수가 호출되었습니다")
    yield "test"

# 함수를 호출한다.
print('A 지점 통과')
test()
print('B 지점 통과')
test()
print(test())
```

```python
## (5) 좀 더 알아보기-제너레이터(Generator)
## 제너레이터 객체와 next() 함수

# 함수를 선언한다.
def test():
    print('A 지점 통과')
    yield 1
    print('B 지점 통과')
    yield 2
    print('C 지점 통과')
# 함수를 호출한다.
output=test()
# next() 함수를 호출한다.
print("D 지점 통과")
a=next(output)
print(a)
b=next(output)
print(b)
print("F지점 통과")
c=next(output) # next() 함수를 호출한 이후 yield 키워드를 만나지 못하고 함수가 끝남 --> StopIteration 에외 발생함.
print(c)
# 한번 더 실행하기
next(output)
```
