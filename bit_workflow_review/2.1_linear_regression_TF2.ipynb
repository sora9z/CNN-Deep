# 2.1 Linear Regression by Tensorflow2


```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

x_train=[1,2,3,4]
y_train=[2,4,6,8]

#model instance
tf.model=Sequential()
#units=output shape , input_dim=input shape
tf.model.add(Dense(units=1,input_dim=1)) #layer

#Set optimizer and compile(loss,optimizer)
optimize=SGD(lr=0.01) #SGD==Standard Gradient Descent ,lr=learning rate
tf.model.compile(loss='mse',optimizer=optimize) #mse==mean_sequared_error,1/m*sgd(y'-y)^2

#Fit() executes training 
tf.model.fit(x_train,y_train,epochs=200)

#pridict() returns predicted value
y_predict=tf.model.predict(np.array([5,4]))
print(y_predict)
```

```python
# 한번 더 연습 (y_train=x_train*5+3)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import numpy as np

# Set Data
x_train=[i for i in range(1,50,2)]
y_train=[5*i+3 for i in x_train]
print(y_train)

# Set Model 
tf.model=Sequential()
# Add layer
tf.model.add(Dense(units=1,input_dim=1))

# Set Optimizer and Train
optimize=SGD(lr=0.001)
tf.model.compile(loss='mse',optimizer=optimize)

# Print summary of the model to the terminal
tf.model.summary()

# excute graph
tf.model.fit(x_train,y_train,epochs=200)

# Prediction
y_predict=tf.model.predict(np.array([i for i in range(1,30)]))
print(y_predict)
```
